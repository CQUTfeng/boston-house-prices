```python
feature = data[['RM','PTRATIO','LSTAT']]
price = data['MEDV']
```

```python
feature['LSTAT'] = feature['LSTAT'].fillna(feature['LSTAT'].mean())
```

```python
feature.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, price, test_size=0.3, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_predict = lr.predict(X_test)
print(r2_score(y_test, y_predict))
```

```python
print(mean_squared_error(y_test, y_predict))
```

```python
mean_squared_error(y_test, y_predict)**0.5
```

```python
from sklearn.learning_curve import learning_curve, validation_curve
```

```python
train_sizes, train_scores, test_scores = learning_curve(lr, feature, price, train_sizes=[0.1,0.2,0.4,0.6,0.8], cv=5)
```

```python
plt.grid()
plt.plot(train_sizes, train_scores.mean(axis=1),'r')
plt.plot(train_sizes, test_scores.mean(axis=1),'g')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import make_scorer
```

```python
def performance_metric(y_predict, y_test):
    score = r2_score(y_predict, y_test)
    return score

def fit_model(X, y):
    reg = DecisionTreeRegressor()
    params = {'max_depth':np.arange(1,11)}
    scoring_fnc = make_scorer(performance_metric)
    cross_validator = KFold()
    grid = GridSearchCV(reg, params, scoring_fnc, cv=cross_validator)
    grid.fit(X, y)
    
    print(pd.DataFrame(grid.cv_results_))
    return grid.best_estimator_
```

```python
optimal_reg = fit_model(X_train, y_train)
print(optimal_reg)
```

```python
y_predict = optimal_reg.predict(X_test)
print(performance_metric(y_test, y_predict))
print(mean_squared_error(y_test, y_predict))
```

```python
reg = DecisionTreeRegressor(max_depth=4)
reg.fit(X_train, y_train)
y_predict1 = reg.predict(X_test)
print(r2_score(y_test, y_predict1))
print(mean_squared_error(y_test, y_predict1))
```
